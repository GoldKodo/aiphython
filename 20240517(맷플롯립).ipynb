```python
#맷플롯립 (MATPLOTLIB)
import matplotlib.pyplot as plt
x = [1,2,3,4] #x축 데이터
y = [1,4,9,16] #y축 데이터
plt.plot(x,y)   #plot(x값, y값, 선스타일문자열, marker=‘데이터포인트마커기호’, label=‘그래프이름'): y값은 필수이고 나머지는 생략 가능
plt.show()
```

```python
#맷플롯립 예제2
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel('data1.xlsx')
plt.plot(df['kor'], '--', marker='o', label='korean') #데이터 프레임의 kor 항목을 y값, 선스타일= --, 마커기호= o, 그래프이름 = korean 으로 해석.
plt.plot(df['math'], marker='^',label='math')
plt.legend()
plt.show()
```

```python
#맷플롯립 실습
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('font', family='NanumGothic')

df=pd.DataFrame({'요일': ['월','화','수','목','금','토','일'],
                 '공부시간': [2,3,1,5,4,2,1],
                 '성적': [70,75,65,85,80,70,60]
                 })

plt.plot(df['공부시간'], '--', marker='o', label='공부시간')
plt.plot(df['성적'],marker='^',label= '성적')
plt.legend()  #이름 들어가는 오른쪽 위 박스 표시
plt.show()
```

```python
#산점도 그래프(scatter)
import matplotlib.pyplot as plot
x = [6,2,1,1,4,4,6,4,7,7]
y = [5,4,2,9,9,0,10,3,1,2]
plt.scatter(x, y) #scatter( x값, y값, c=‘색 문자’ 혹은 ‘색깔 문자열’, marker=‘마커기호’, label=‘그래프이름’ ): c, marker, label은 생략 가능
plt.show()
```

```python
#바 그래프
import matplotlib.pyplot as plt
x = ['A','B','C']
y = [100, 150, 120]
plt.bar(x,y) #bar(x값, y값, color=‘색 문자’ 혹은 ‘색깔 문자열’) : 범주별 데이터를 나타내는 세로 막대 그래프 생성 (barh는 가로 막대 그래프)
plt.title('Test Graph') #그래프 제목
plt.xlabel('Group') #x축 제목
plt.ylabel('Score') #y축 제목
plt.show()
```

```python
#막대 그래프 실습
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')

x = ['서울','부산','인천','대구']
y = [9904312,3448737,2890451,2466052]

plt.bar(x,y)
plt.title('도시별 인구 수')
plt.xlabel('도시')
plt.ylabel('인구 수')
plt.show()
```

```python
#한화면에 그래프 여러개 만들기
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2) #subplots(행 개수, 열 개수, figsize=(가로 길이, 세로 길이)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

```python
#데이터1 사용 선그래프 그리기
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('font', family='NanumGothic')

df = pd.read_excel('data1.xlsx')
plt.plot(df['name'],df['kor'],'g--',marker = 'o', label = '국어 점수')
plt.plot(df['name'],df['math'],'r',marker = 'v', label = '수학 점수')
plt.title('성적 그래프')
plt.xlabel('이름')
plt.ylabel('점수')
plt.legend()
plt.show()
```
