```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.info()
#종별 개수 확인
df종수 = df['species'].value_counts()
print(df종수)
#결측치 확인
df.isnull().sum()
df.isna().sum()
#중복값 확인
df[df.duplicated()]
#이후 중복이 나오면
idx = (df['sepal_length'] ==5.8) & (df['petal_width'] == 1.9)
df.loc[idx,:]  #각 값이 몇인지 확실히 하고 찍으면 중복값의 위치를 알 수 있음
#중복값 처리
df2= df.drop_duplicates()
#열 기준 합계 / 평균
df2.groupby('species').sum()
df2.groupby('species').mean()
#막대 그래프 그리기
import matplotlib.pyplot as plt
plt.bar(df.index[:10], df.sepal_length[:10])
plt.xlabel('data')
plt.ylabel('sepal_length')
plt.show()
#품종별 막대 그래프
import seaborn as sns
df3 = df2.groupby('species').mean()
sns.barplot(x = 'species', y = 'sepal_length', data = df3)
plt.show()

#히스토그램그래프
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    plt.hist(df[feature], bins=20, alpha = 0.7)
plt.show()

#꽃받침의 길이와 꽃잎의 길이의 산점도를 시각화
sns.scatterplot(x='petal_length',y='sepal_length',hue='species',data= df2)
plt.show()
```
