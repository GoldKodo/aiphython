```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.info()
#  
df = df['species'].value_counts()
print(df)
# 
df.isnull().sum()
df.isna().sum()
# 
df[df.duplicated()]
#  
idx = (df['sepal_length'] ==5.8) & (df['petal_width'] == 1.9)
df.loc[idx,:]  #          
# 
df2= df.drop_duplicates()
#   / 
df2.groupby('species').sum()
df2.groupby('species').mean()
#  
import matplotlib.pyplot as plt
plt.bar(df.index[:10], df.sepal_length[:10])
plt.xlabel('data')
plt.ylabel('sepal_length')
plt.show()
#  
import seaborn as sns
df3 = df2.groupby('species').mean()
sns.barplot(x = 'species', y = 'sepal_length', data = df3)
plt.show()

#
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    plt.hist(df[feature], bins=20, alpha = 0.7)
plt.show()

#     
sns.scatterplot(x='petal_length',y='sepal_length',hue='species',data= df2)
plt.show()
```
