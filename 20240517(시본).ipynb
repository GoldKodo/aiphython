```python
#시본으로 산점도 그래프 그리기: scatterplot()
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

iris_data=pd.read_csv('iris.csv')
sns.scatterplot(x='petal_length',y='petal_width',hue='species',data= iris_data)
#scatterplot(x=‘x축에 표시할 칼럼명＇, y=‘y축에 표시할 칼럼명’, hue=‘데이터구분 칼럼명‘, data=데이터프레임 ): hue값 지정은 생략 가능
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
#실습
dt = pd.DataFrame({
'x': [1, 2, 3, 4, 5],
'y': [10, 20, 15, 25, 30],
'category': ['A', 'B', 'A', 'B', 'A']
})

sns.scatterplot(x= 'x',y= 'y', hue='category',data= dt)
```

```python
#범주별 데이터 개수를 표시하는 막대 그래프: counterplot()

data = pd.DataFrame({
'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],
'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]
})

sns.countplot(x= 'category', hue= 'category', data=data)
#•countplot( x=‘x축에 범주로 표시할 칼럼명＇, hue=‘데이터 구분 칼럼명‘, data=데이터프레임 ): hue값 지정은 생략 가능
plt.show()
```

```python
# 범주별 항목값 평균을 표시하는 막대 그래프 : barplot( )
# barplot( x=‘x축에 표시할 칼럼명＇, y=‘y축에 표시할 칼럼명’, data=데이터프레임, hue ='데이터 구분 칼럼명')
# 숫자값을 갖는 칼럼들의 상관관계를 보여주는 그래프 : pairplot( )
# pairplot(데이터프레임, hue=‘범주를 갖는 칼럼명＇)
# 산점도와 추세선(회귀선)을 동시에 보여주는 그래프 : regplot( )
# regplot(x=‘x축에 표시할 칼럼명‘, y=‘y축에 표시할 칼럼명‘,data= 데이터프레임)
# 집계한 값에 따라 색깔을 표시하는 그래프 : heatmap( )
# heatmap(data=2차원 행렬 형태로 전달된 데이터, annot=True/False, fmt=‘파이썬 서식 문자열‘, cbar=True/False)

pd=sns.load_dataset('penguins')
pd
# 1.데이터를 가져온 후 펭귄 품종(species)별 몸무게(body_mass_g)의 평균을  성별을 구분해서 그래프로 그려라
sns.barplot(x= 'species', y= 'body_mass_g',data=pd, hue = 'sex')
# 2.penguins 데이터의 숫자값을 갖는 항목들끼리의 상관관계 그래프를 그려보자
sns.pairplot(pd, hue= 'species')
# 3.2번 미션의 결과에서 칼럼들간의 상관관계를 확인하고 관계가 있는 두 칼럼을 선택하여 산점도와 추세선이 함께 나타나는 그래프를 그려보자.
sns.regplot(x= 'body_mass_g', y= 'bill_length_mm', data=pd)
```
